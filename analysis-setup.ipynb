```python
import sqlite3
from sqlite3 import Error
import yaml
from pathlib import Path
import os
from  parse_webui import parse_result_files
```

```python
def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection
```

```python
conn = create_connection('database.db')
```

## get images

```python
def write_to_file(data, filepath):
    # Convert binary data to proper format and write it on Hard Disk
    with open(filepath, 'wb') as file:
        file.write(data)
    print("Stored blob data into: ", filepath, "\n")
```

```python
cur = conn.cursor()
select_all_images_query = "SELECT image, filename FROM prompts"
cur.execute(select_all_images_query)
record = cur.fetchall()
cur.close()
for (image, filename) in record:
    write_to_file(image, f"./images/txt2img/{filename}.png")
```
